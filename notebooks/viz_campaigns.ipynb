# Campaigns / Categories Comparison

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../campaigns/simulated_ads.csv')
df.head()
```

```python
agg = df.groupby(['category', 'month'])[['spend','leads']].sum().reset_index()
for metric in ['spend','leads']:
    plt.figure()
    for cat, sub in agg.groupby('category'):
        sub = sub.sort_values('month')
        plt.plot(sub['month'], sub[metric], marker='o', label=cat)
    plt.title(f'{metric.title()} by Category')
    plt.xlabel('Month')
    plt.ylabel(metric.title())
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
