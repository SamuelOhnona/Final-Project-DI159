# Performance Visualization
Matplotlib-only charts of monthly KPIs from `campaigns/simulated_ads.csv`.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../campaigns/simulated_ads.csv')
df.head()
```

```python
plt.figure()
df.groupby('month')[['leads']].sum().plot(marker='o')
plt.title('Leads per Month')
plt.xlabel('Month')
plt.ylabel('Leads')
plt.show()
```

```python
plt.figure()
df.groupby('month')[['cpc']].mean().plot(marker='o')
plt.title('Average CPC per Month')
plt.xlabel('Month')
plt.ylabel('CPC ($)')
plt.show()
```

```python
plt.figure()
df.groupby('month')[['cpl']].mean().plot(marker='o')
plt.title('Average CPL per Month')
plt.xlabel('Month')
plt.ylabel('CPL ($)')
plt.show()
```
